```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("updates_for_training_ml.csv")
df["score"] = None
df.head()
```

```python
import joblib

# Charger votre modèle sauvegardé
# Remplacez 'nom_de_votre_modele.pkl' par le chemin réel de votre fichier
model = joblib.load('model.pkl')
# --- Préparation des colonnes ---
# Récupérer la liste des features attendues par le modèle
# Cette ligne ne fonctionne que si votre modèle est un pipeline scikit-learn entraîné sur un DataFrame pandas
try:
    features_attendues = model.feature_names_in_
    # Créer un DataFrame avec uniquement les colonnes attendues, dans le bon ordre
    df_pour_prediction = df[features_attendues]
except AttributeError:
    # Si votre modèle n'a pas .feature_names_in_, vous devez définir la liste manuellement
    print("Le modèle ne contient pas 'feature_names_in_'. Vous devez définir manuellement les colonnes.")
    # Exemple : features_attendues = ['col1', 'col2', 'col3', ...]
    # df_pour_prediction = df[features_attendues]
    # Pour l'instant, on suppose que le df a les bonnes colonnes
    df_pour_prediction = df 
```

```python
probabilites = model.predict_proba(df_pour_prediction)
```

```python
score_classe_positive = probabilites[:, 1]
```

```python
df['score'] = score_classe_positive
```
