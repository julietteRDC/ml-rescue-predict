```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from geopy.distance import great_circle
import plotly.express as px
```

```python
caract_df = pd.read_csv("data/caract-2023.csv", delimiter=";")
lieux_df = pd.read_csv("data/lieux-2023.csv", delimiter=";")
usagers_df = pd.read_csv("data/usagers-2023.csv", delimiter=";")
vehicules_df = pd.read_csv("data/vehicules-2023.csv", delimiter=";")
```

```python
# ⚡ Paramètres
EPS_METERS = 5000  # 5 km
EPS_KM = EPS_METERS / 1000  # Conversion en km

# Passer les latitudes et longitudes en float
caract_df[['lat', 'long']] = caract_df[['lat', 'long']].replace(',', '.', regex=True).astype(float)
caract_df[['lat', 'long']] = caract_df[['lat', 'long']].astype(float)

# Conversion en tableau numpy
coords = caract_df[['lat', 'long']].to_numpy()

# ⚡ Clustering avec DBSCAN (Distance en km)
db = DBSCAN(eps=EPS_KM / 6371, min_samples=2, metric='haversine').fit(np.radians(coords))

# Ajouter les clusters aux données
caract_df['cluster'] = db.labels_

# 💡 Vérification des clusters
print(caract_df['cluster'].value_counts())

# 📌 Détermination du centre de la carte
center_lat, center_lon = caract_df[['lat', 'long']].mean()

# 📌 Affichage des clusters sur une carte
fig = px.scatter_mapbox(caract_df, 
                        lat="lat", 
                        lon="long", 
                        text="adr", 
                        hover_name="adr", 
                        color="cluster", 
                        zoom=8, 
                        mapbox_style="carto-positron")

# Ajustement du centre
fig.update_layout(mapbox_center={"lat": center_lat, "lon": center_lon})

fig.show()
```
